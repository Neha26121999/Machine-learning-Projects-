# Step 4: Split dataset into train and test model

```python
from sklearn.model_selection import train_test_split
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_train_sc=scale.fit_transform(x_train)
x_test_cs=scale.transform(x_test)
```

# Step 5: Apply Machine Learning Models

1. This is with scaled data

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train_sc,y_train)
```

```python
y_pred=svc.predict(x_test_cs)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,f1_score
accuracy_score(y_test,y_pred)
```

2. Now this is without scaled data

```python
from sklearn.svm import SVC
col=SVC()
col.fit(x_train,y_train)
y_pred_sc=col.predict(x_test)
accuracy_score(y_test,y_pred_sc)
```

Algo 2: without Scaled data

```python
from sklearn.linear_model import LogisticRegression
lol=LogisticRegression(random_state = 51, penalty = 'l1')
lol.fit(x_train,y_train)
y_pred=lol.predict(x_test)
accuracy_score(y_test,y_pred)
```

LogisticRegression with sacled data

```python
from sklearn.linear_model import LogisticRegression
lol=LogisticRegression(random_state = 51, penalty = 'l1')
lol.fit(x_train_sc,y_train)
y_pred_ll=lol.predict(x_test_cs)
accuracy_score(y_test,y_pred_ll)
```

```python
confusion_matrix(y_test,y_pred_ll)
```
